```python
import matplotlib.pyplot as plt
```

```python
x = 2.0
lr = 0.1
x_values = []
y_values = []
```

```python
def cost_function(x):
    return (x+3)**2

def gradient(x):
    return 2*(x + 3)

    
```

```python
no_iterations = 100
```

```python
for i in range(no_iterations):
    x_values.append(x)
    y_values.append(cost_function(x))
    grad = gradient(x)
    x = x - lr * grad
    # print(f"Iteration {i+1:3d}: x = {x:.6f}, Cost = {cost_function(x):.10f}")
    print(f"Iteration {i+1}: x = {x:.6f}, Cost = {cost_function(x):.10f}")


# Step 4: Final result
print(f"\n✅ Optimal x (Local Minima): {x:.6f}")
```

```python
plt.plot(x_values , y_values , 'ro',linestyle = '-',linewidth = 2)
plt.title("Gradient descent for given eq")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
