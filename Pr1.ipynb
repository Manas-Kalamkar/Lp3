```python
nterms = 6

n1,n2 = 0,1;
count = 0;

if nterms<=0 :
    print("Please enter the positive number");

elif nterms == 1:
    print("Fibonacci sequence upto ",nterms,": ")
    print(n1);

else:
    print("Fibonacci sequence: ")
    while count < nterms:
        print(n1);
        nth = n1 + n2;

        n1 = n2
        n2 = nth
        count +=1
```

```python
def recur_fibo(n):
    if n<=1 :
        return n
    else:
        return (recur_fibo(n-1) + recur_fibo(n-2))

nterms = int(input("Enter the number till you want fibonacci sequence: "));
if(nterms <=0):
    print("please enter positive number");
else:
    print("Fibonacci sequence: ");
    for i in range (nterms):
        print(recur_fibo(i))
```
